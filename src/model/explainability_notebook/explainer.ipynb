# Importando as libs

Aqui importo o tokenizer, torch para importar os pesos do modelo treinado e a classe Explainer criado com SHAP

```python
import tiktoken
import torch
from shap_explainer import Explainer
from ..gpt2 import GPT, GPTConfig

model = GPT(GPTConfig(block_size=256, vocab_size=50304, n_layer=4, n_head=4, n_embd=256))
weights_path = "../weights/gpt2_weights.pth"
enc = tiktoken.get_encoding("gpt2")

model.load_state_dict(torch.load(weights_path))
```

## Aqui chamarei a classe de explicabilidade 

```python
explainer = Explainer(model, enc)
```

A classe acima ira explicar o que cada token de input ira impactar na parte autoregressiva do modelo, o output sera um grafico <br>
que por token explicara o output token a token, o codigo abaixo exemplificara melhor 

```python
explainer.explain("Scientists confirmed the worst possible outcome")
```

Aqui e exemplificado cada modificacao do logit token a token para mostrar o que ocorre

## In depth do que esta acontecendo

A cada token processado e gerado o logit sera alterado mesmo apos o input, entao o shap nos provisiona um metodo de como <br>
ocorre a subida do gradiente pela probabilidade dada

## Explicando a classe Explainer

```python
import shap


class Explainer:
    def __init__(self, model, tokenizer):
        self.explainer = shap.Explainer(model, tokenizer)

    def explain(self, text):
        return self.explainer(text)
```

Aqui temos o chamado do objeto Explainer do shap que ira retornar o logit passado e como ele altera os resultados e nao <br>
necessita ser recarregado a cada run de input e basta chamar o explain com o texto input
